```python
import pandas as pd
```

```python
planilha = pd.read_html('https://docs.google.com/spreadsheets/d/e/2PACX-1vSUuul2dZjGK68GSyxdTklLAA4oS_7_eiB5iHv5H9GlZkOWO-BBAg-fLtRr-cuRNKPYYPCqXNH0bjxp/pubhtml')
```

```python
planilha
```

```python
planilha = planilha[0][1:]
```

```python
planilha
```

```python
planilha.rename({'Unnamed: 9' : 'Telefones'}, axis = 1, inplace = True)
```

```python
planilha
```

```python
planilha['Telefones_Corretos'] = planilha.Telefones.apply(lambda x: str(x))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.rstrip())
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.lstrip())
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace('(',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace(')',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace('())',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace(' ',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace('-',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace('+',''))
planilha['Telefones_Corretos'] = planilha.Telefones_Corretos.apply(lambda x: x.replace('+55',''))
```
